```python
data.train=pd.read_csv("C:\\Users\\Vitya_\YandexDisk\ML coursera\RAN\week2_knn\perceptron-train.csv",header=None)
data.test=pd.read_csv("C:\\Users\\Vitya_\YandexDisk\ML coursera\RAN\week2_knn\perceptron-test.csv",header=None)
data.test.head()
```
