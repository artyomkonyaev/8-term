```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("abalone.csv")
df
```

```python
df['Sex'] = df['Sex'].map(lambda x: 1 if x == 'M' else (-1 if x == 'F' else 0))
```

```python
df.head()
```

```python
y=df.Rings
del df['Rings']
X=df
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import KFold,cross_val_score
from sklearn import grid_search
```

```python
%%time
grid={'n_estimators':np.arange(1,51)}
cv=KFold(y.size,n_folds=5,shuffle=True,random_state=1)
rgr=RandomForestRegressor(random_state=1)
gs=grid_search.GridSearchCV(rgr,grid,scoring='r2',cv=cv)
gs.fit(X,y)
```

```python
for a in gs.grid_scores_:
    print(a.mean_validation_score,a.parameters)
```
